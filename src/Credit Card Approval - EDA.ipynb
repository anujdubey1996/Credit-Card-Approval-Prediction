## Credit Card Approval

### Business Understanding:
For many banks, retaining high profitable customers is the number one business goal. Interest from EMIs and merchant fee make credit cards one of the main source of income for them. But just like a simple loan, credit card defaulters pose significant threat to this goal for banks. In terms of substantial financial losses, trust and credibility, this is a concerning issue to both banks and their customers.
<br>
<br>
**Understanding the Credit Card approval**:
 To make sure that credit card default is avoided without affecting the income from interests, the credit card should be approved for a customer with a credit limit he/she can repay. To ensure that various factors can be considered:
- Credit History
- Limits from other credit cards
- Occupation
- Marital Status

### Data understanding:
The data set includes details of bank's existing users and whether or not they were approved for a credit card. Out 45222 users, 11208 users were approved for the card. This dataset also contains their personal information like their gender, occupation, marital status, education level and their credit card history.
